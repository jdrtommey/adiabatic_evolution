```python
import hohi
import numpy as np
```

```python
h0 = np.diag([-2,-1,0,1,2])
```

```python
h1 = np.random.rand(5,5)
for i in range(h1.shape[0]):
    h1[i][i] = 0.0
```

```python
my_solver = hohi.Adiabatic_solver(h0,h1,np.linspace(0,5,1000),0.1,5)
success = my_solver.run()
```

```python
import matplotlib.pyplot as plt
```

```python
success[0].get_length()
```

```python
fig = plt.figure()
ax = fig.add_subplot(121)
ax1 = fig.add_subplot(122)
for s in success:
        ax.scatter(s.parameter,abs(s.vecs[:,s.index]),s=0.1)
        
for s in success:
        ax1.scatter(s.parameter,s.vals,s=0.1)
```
