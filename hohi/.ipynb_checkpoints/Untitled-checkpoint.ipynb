```python
from scipy.sparse import csr_matrix
import hohi
import numpy as np
import matplotlib.pyplot as plt
plt.style.use('ggplot')
```

```python
h0 = np.diag([-3.0,-2.0,-1.0,0.01,1,2,3])
type(h0)
```

```python
h1 = np.zeros((7,7))
h1[0][6] = 0.5
h1[6][0] = 0.5

h1[2][3] = 0.1
h1[3][2] = 0.1

h1[4][2] = 0.1
h1[2][4] = 0.1

h1[5][3]=0.1
h1[3][5]= 0.1

h1[1][2] = 0.5
h1[2][1] = 0.5

```

```python
solver = hohi.Adiabatic_solver(h0,h1,np.linspace(0.0,8,1000),-1.012,5,sig_method='target',target=[1],return_failed=True,index_list=[0,1,2,3,4])
```

```python
success,fail = solver.run()
```

```python
for state in success:
    plt.scatter(state.parameter,state.vals,s=1)

```

```python
for state in success:
    plt.scatter(state.parameter,abs(state.vecs[:,state.index]),s=1)
```

```python
for state in success:
    plt.scatter(state.parameter,abs(state.vecs[:,success[2].index]),s=1)
```
